```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score  
```

```python
# Step 1: Load the dataset
data = pd.read_csv(r"C:\Users\sowmy\Downloads\house(1).csv")
# Step 2: Check for missing values
data.isnull().sum()
# Step 3: Drop the 'sqft_above' column
data.drop(['sqft_above'],axis = 'columns',inplace = True)
# Step 4: Convert the 'date' column to binary values based on the year (2014 or not)
con_date=[1 if values == 2014 else 0 for values in data.date]
data['date']=con_date
```

```python
# Step 5: Inspect the first few rows of the cleaned data
data.head()
```

```python
# Step 6: Get information about the dataset, including column types and non-null counts
data.info()
```

```python
 # Target variable (house price)
y = data[['price']].values
# Features (all columns except 'id' and 'price')
x=data.drop(['id','price'], axis = 1).values
```

```python
#Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
#Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#Make predictions using the test data
y_predict = model.predict(x_test)
# Print predicted and actual values for comparison
print(y_predict)
print(y_test)
```

```python
# Step 6: Evaluate the model using the coefficient of determination (R² score)
print("coefficient of determination: %.4f" % r2_score(y_test,y_predict))
```
